```python
'''
Question 3
Which days of the week have the highest number of bookings?
Also, draw a pie chart to show the distribution for all days of the week?
'''
import matplotlib.pyplot as plt
# Extract the day of the week from booking_time
Bookings_df['day_of_week'] = Bookings_df['booking_time'].dt.day_name()
# Count the number of bookings for each day of the week
Bookings_per_day = Bookings_df['day_of_week'].value_counts()
# Find the day with the highest number of bookings
day_with_highest_bookings = Bookings_per_day.idxmax()
print("\033[2mDay with the highest number of bookings is:\033[1m", day_with_highest_bookings)

# Plot a pie chart for the distribution of bookings across days of the week
plt.figure(figsize=(10, 6))
Bookings_per_day.plot.pie(autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired(range(7)))
plt.title('Distribution of Bookings Across Days of the Week')

# Hide the y-label for better aesthetics
plt.ylabel('')
plt.show()
```

```python
'''
Question 4
For each of the service names, display the total number of bookings and the total Gross Booking Value in INR.
'''
# Group bookings by service_name
Service_Stats = Bookings_df.groupby('service_name').agg(
    total_bookings=('booking_id', 'count'),
    Total_gross_booking_value=('INR_Amount', 'sum')
).reset_index()
print("\033[2mThe Number of Total Bookings and Gross Booking Value are:\033[1m")
Service_Stats_reset = Service_Stats.reset_index(drop=True)
Service_Stats_reset.index = Service_Stats_reset.index + 1
Service_Stats_reset
```

```python
'''
Question 5
For customers who have more than 1 booking, which is the most booked route (from_city to to_city)?
'''
# Filter customers who have more than 1 booking
Customer_booking_counts = Bookings_df.groupby('customer_id')['booking_id'].count()
Customers_with_multiple_bookings = Customer_booking_counts[Customer_booking_counts > 1].index
# Filter the bookings data to include only these customers
Multiple_bookings_df = Bookings_df[Bookings_df['customer_id'].isin(Customers_with_multiple_bookings)]
# Find the most booked route for these customers
Route_Counts = Multiple_bookings_df.groupby(['from_city', 'to_city']).size().reset_index(name='count')
Most_Booked_Route = Route_Counts.loc[Route_Counts['count'].idxmax()]
print("\033[2mMost Booked Route for customers with more than 1 booking:\033[1m")
print(f" The Most Route Booking is From {Most_Booked_Route['from_city']} to {Most_Booked_Route['to_city']} with {Most_Booked_Route['count']} bookings")
```

```python
'''
Question 6
Which are the top 3 departure cities from where customers book mostly in advance, provided that there have been at least 5 departures from that city?
'''
# Filter bookings where days_to_departure > 0
Advance_Bookings_df = Bookings_df[Bookings_df['days_to_departure'] > 0]
# Group by from_city and calculate the number of advance bookings and average days to departure
City_Advance_Stats = Advance_Bookings_df.groupby('from_city').agg(
    Total_Departures=('booking_id', 'count'),
    Average_Days_to_Departure=('days_to_departure', 'mean')
).reset_index()
# Filter cities with at least 5 departures
City_Advance_Stats_Filtered = City_Advance_Stats[City_Advance_Stats['Total_Departures'] >= 5]
# Sort by average days to departure in descending order and select the top 3 cities
top_3_cities = City_Advance_Stats_Filtered.sort_values(by='Average_Days_to_Departure', ascending=False).head(3)
print("\033[2mTop three departure cities with the most advance bookings (at least 5 departures) are:\033[1m")
# Reset the index
top_3_cities_reset = top_3_cities.reset_index(drop=True)
# Change the index to start from 1
top_3_cities_reset.index = top_3_cities_reset.index + 1
top_3_cities_reset
```

```python
# Also Checking for the top most departing cities by total departures in descending order

Top_Three_Cities = top_3_cities_reset.sort_values(by='Total_Departures', ascending=False)
Top_Three_Cities_reset = Top_Three_Cities.reset_index(drop=True)
Top_Three_Cities_reset.index = Top_Three_Cities_reset.index + 1

Top_Three_Most_Departing_Cities = Top_Three_Cities_reset

print("Top three most departing cities with the most total departures are:")
Top_Three_Most_Departing_Cities
```

```python
'''
Question 7
Plot a heatmap displaying correlations of the numerical column and report which pair of numerical columns in the bookings data set, have the maximum correlation?
'''
import seaborn as sns # for data visualization
import matplotlib.pyplot as plt
# Selecting a numerical columns for correlation analysis in Bookings Data Frame
numerical_columns = Bookings_df.select_dtypes(include='number').columns
# Computing the correlation matrix by numerical column
correlation_matrix = Bookings_df[numerical_columns].corr()
# Making a heatmap plot from the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix of Numerical Columns in Bookings Data')
plt.show()

# Find the pair of numerical columns with the maximum correlation
correlation_unstacked = correlation_matrix.unstack()
# Exclude self-correlation by filtering out 1.0 and taking the absolute value for finding the max correlation
correlation_unstacked = correlation_unstacked[correlation_unstacked < 1].abs()
max_correlation_pair = correlation_unstacked.idxmax()
max_correlation_value = correlation_unstacked.max()

print(f"\033[2mThe pair of numerical columns with the maximum correlation is:\033[1m {max_correlation_pair} with a correlation of {max_correlation_value:.2f}")
```

```python
'''
Question 8
For each service, which is the most used device type for making bookings on the platform?
'''
# Firstly you group bookings by service_name and device_type_used and count the number of bookings for each combination
Device_Usage_Stats = Bookings_df.groupby(['service_name', 'device_type_used']).size().reset_index(name='count')
# Find the most used device type for each service
Most_Used_Device_by_Service = Device_Usage_Stats.loc[Device_Usage_Stats.groupby('service_name')['count'].idxmax()]
print("\033[2mMost used device type for each service:\033[1m")

Most_Used_Device_by_Service_reset = Most_Used_Device_by_Service.reset_index(drop=True)

Most_Used_Device_by_Service_reset.index = Most_Used_Device_by_Service_reset.index + 1

Most_Used_Device_by_Service_reset
```

```python
'''
Question 9
Plot the trends at a quarterly frequency for the number of bookings by each of the device types, that is, plot a time series for each year and quarter showing the number of bookings performed by each device type.
'''
# Firstly, Confirm the Booking_time is in datetime format
Bookings_df.dtypes
```

```python
# Secondly, Set booking_time as the index, since booking_is already in datetime format, it is ready to set the dataframe's index to a specified column
Bookings_df.set_index('booking_time', inplace=True)
```

```python
# Resample the data quarterly and count the number of bookings for each device type, thereby aggregating the data for each quarter
Quarterly_Device_Bookings = Bookings_df.groupby('device_type_used').resample('Q').size().unstack(level=0, fill_value=0)
```

```python
# Plot the time series for each device type
plt.figure(figsize=(12, 8))
for device_type in Quarterly_Device_Bookings.columns:
    plt.plot(Quarterly_Device_Bookings.index, Quarterly_Device_Bookings[device_type], marker='o', label=device_type)

plt.title('Quarterly Number of Bookings by Device Type')
plt.xlabel('Quarter')
plt.ylabel('Number of Bookings')
plt.legend(title='Device Type')
plt.grid(False)
plt.show()
```

```python
'''
Question 10
Consider the following example:12 customers performed a total of 100 searches but only a few of them performed 10 bookings in total.
Searches = 100 bookings = 10 customers ,1 , The overall booking to search rati, Obe, is 10/100 = 0.

1 Using the above information/context, answer the follongwi
• What is the average oBSR for each month of the year?
• What is the average oBSR for each day of the week?
• _Plot a time series of oBSR on all the given datess.
'''
import pandas as pd
import matplotlib.pyplot as plt
# First, we will calculate the average oBSR for each month and day of the week by merging sessions and bookings dataframe using a leftjoin
merged_sessions_bookings = pd.merge(Sessions_df, Bookings_df, on="booking_id", how="left")
# Secondly, remove timezone information from the search_time column
merged_sessions_bookings['search_time'] = merged_sessions_bookings['search_time'].dt.tz_localize(None)
# Thirdly, calculate the number of searches and bookings per month
searches_per_month = merged_sessions_bookings.groupby(merged_sessions_bookings['search_time'].dt.to_period('M')).size()
bookings_per_month = merged_sessions_bookings.dropna(subset=['booking_id']).groupby(merged_sessions_bookings['search_time'].dt.to_period('M')).size()
# Next is to calculate oBSR per month
obsr_per_month = bookings_per_month / searches_per_month
obsr_per_month = obsr_per_month.fillna(0)  # Fill NaN values with 0
# Calculate the number of searches and bookings per day of the week
searches_per_day = merged_sessions_bookings.groupby(merged_sessions_bookings['search_time'].dt.day_name()).size()
bookings_per_day = merged_sessions_bookings.dropna(subset=['booking_id']).groupby(merged_sessions_bookings['search_time'].dt.day_name()).size()
# Calculate oBSR per day of the week
obsr_per_day = bookings_per_day / searches_per_day
obsr_per_day = obsr_per_day.fillna(0)  # Fill NaN values with 0
# Calculate the number of searches and bookings per day
searches_per_day_series = merged_sessions_bookings.groupby(merged_sessions_bookings['search_time'].dt.date).size()
bookings_per_day_series = merged_sessions_bookings.dropna(subset=['booking_id']).groupby(merged_sessions_bookings['search_time'].dt.date).size()
# Calculate oBSR per day
obsr_per_day_series = bookings_per_day_series / searches_per_day_series
obsr_per_day_series = obsr_per_day_series.fillna(0)  # Fill NaN values with 0
# Plot the oBSR per month
plt.figure(figsize=(12, 6))
obsr_per_month.plot(marker='o', linestyle='-', title='Average oBSR per Month')
plt.ylabel('oBSR')
plt.xlabel('Month')
plt.grid(False)
plt.show()
```

```python
# Plot the oBSR per day of the week
plt.figure(figsize=(12, 6))
obsr_per_day.plot(kind='bar', title='Average oBSR per Day of the Week')
plt.ylabel('oBSR')
plt.xlabel('Day of the Week')
plt.grid(False)
plt.show()
```

```python
# Plot the time series of oBSR per day
plt.figure(figsize=(12, 6))
obsr_per_day_series.plot(marker='o', linestyle='-', title='Time Series of oBSR per Day')
plt.ylabel('oBSR')
plt.xlabel('Date')
plt.grid(False)
plt.show()
```
